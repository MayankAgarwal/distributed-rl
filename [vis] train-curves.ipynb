```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import os
from scipy.signal import savgol_filter
%matplotlib inline
```

```python
result_folder = "results/"
experiments = filter(lambda x: not x.startswith('.') and not x.startswith('[T]'), os.listdir(result_folder))
rewards_file = 'game_rewards.npy'
results = {}
```

```python
for exp in experiments:
    rewards_filepath = os.path.join(result_folder, exp, rewards_file)
    results[exp] = np.load(rewards_filepath)
```

```python
for exp, rewards in results.iteritems():
    lives = range(1, len(rewards)+1)
    reward_smoothed = savgol_filter(rewards, 151, 2)
    
    plt.figure(figsize=(6, 6))
    plt.plot(lives, reward_smoothed, 'b')
    plt.plot(lives, rewards, 'g', alpha=0.3)
    plt.xlabel('Episodes')
    plt.ylabel('Total reward')
#     plt.xlim((0, 2500))
#     plt.ylim((0, 13))
    plt.title(exp)
```
